```python
import random

# 1. Function to generate random numbers or words and write to a file
def generate_random_file(filename, number_of_lines):
    words = ['apple', 'banana', 'cherry', 'orange', 'grape']
    with open(filename, 'w') as file:
        for _ in range(number_of_lines):
            if random.choice([True, False]):  # 50:50 choice
                file.write(f"{random.randint(1, 1000)}\n")
            else:
                file.write(f"{random.choice(words)}\n")

# 2. Generate file_1.txt with 25 lines
generate_random_file("file_1.txt", 25)

# 3. Read and print every line from a file
def read_and_print_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            print(line.strip())

# 4. Return the number of lines in a file
def count_lines_in_file(filename):
    with open(filename, 'r') as file:
        return sum(1 for _ in file)

# 5. Copy contents from a source file to a destination file
def copy_file(source, destination):
    with open(source, 'r') as src, open(destination, 'w') as dst:
        for line in src:
            dst.write(line)

# 6. Calculate sum of numeric values only from a file
def sum_numeric_lines(filename):
    total = 0
    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            if line.isdigit():
                total += int(line)
    return total

# 7. Print dashboard: number of lines, words, characters
def file_dashboard(filename):
    lines = 0
    words = 0
    characters = 0
    with open(filename, 'r') as file:
        for line in file:
            lines += 1
            words += len(line.split())
            characters += len(line)
    print(f"Dashboard for '{filename}':")
    print(f"  Lines     : {lines}")
    print(f"  Words     : {words}")
    print(f"  Characters: {characters}")

# 8. Parse log file and return dictionary of categories (e.g., INFO)
def log_file_summary(filename):
    summary = {}
    with open(filename, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) >= 3:
                label = parts[2].replace(':', '')
                summary[label] = summary.get(label, 0) + 1
    return summary
```
