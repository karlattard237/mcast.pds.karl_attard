#### Part 1

```python
# 1. Print numbers 1 to 10 using a for loop
print("Numbers from 1 to 10 (for loop):")
for i in range(1, 11):
    print(i)

# 2. Use a while loop to print numbers from 5 to 10
print("\nNumbers from 5 to 10 (while loop):")
x = 5
while x <= 10:
    print(x)
    x += 1

# 3. Check if a number is even or odd
num = 7  # You can change this or make it user input
if num % 2 == 0:
    print(f"\n{num} is even")
else:
    print(f"\n{num} is odd")

# 4. List of 5 fruits and print in uppercase
fruits = ["apple", "banana", "mango", "grape", "kiwi"]
print("\nFruits in uppercase:")
for fruit in fruits:
    print(fruit.upper())

# 5. Print every even number in a list
numbers = [3, 4, 7, 10, 12, 15, 18]
print("\nEven numbers in the list:")
for n in numbers:
    if n % 2 == 0:
        print(n)

# 6. Use while loop with input to collect 3 valid names
valid_names = []
print("\nEnter 3 valid names (non-empty):")
while len(valid_names) < 3:
    name = input(f"Enter name {len(valid_names)+1}: ").strip()
    if name != "":
        valid_names.append(name)
    else:
        print("Name cannot be empty.")

print("Collected names:", valid_names)

# 7. if-elif-else to determine grade from a given mark
mark = int(input("\nEnter a mark (0–100): "))
if 90 <= mark <= 100:
    print("Grade: A")
elif 80 <= mark < 90:
    print("Grade: B")
elif 70 <= mark < 80:
    print("Grade: C")
elif 60 <= mark < 70:
    print("Grade: D")
elif 0 <= mark < 60:
    print("Grade: F")
else:
    print("Invalid mark. Please enter a value between 0 and 100.")
```

#### Part 2

```python
import random

# 1. Dictionary of names and ages – print names of people older than 18
people = {"Alice": 22, "Bob": 17, "Charlie": 19, "Diana": 16, "Edward": 25}
print("People older than 18:")
for name, age in people.items():
    if age > 18:
        print(name)

# 2. Nested for loop – multiplication tables from 3 to 5
print("\nMultiplication tables for 3 to 5:")
for i in range(3, 6):
    print(f"\nTable for {i}:")
    for j in range(1, 11):
        print(f"{i} x {j} = {i * j}")

# 3. While loop with break – stops after 7 iterations
print("\nWhile loop with break after 7 iterations:")
counter = 0
while True:
    print("Iteration:", counter + 1)
    counter += 1
    if counter == 7:
        break

# 4. For loop and if – print all elements > 10
values = [4, 12, 7, 18, 3, 25, 9]
print("\nElements greater than 10:")
for val in values:
    if val > 10:
        print(val)

# 5. Set of 5 random numbers – print all squared
print("\nRandom numbers squared:")
random_numbers = set()
while len(random_numbers) < 5:
    random_numbers.add(random.randint(1, 100))

for num in random_numbers:
    print(f"{num} squared is {num ** 2}")

# 6. Count how many times 'a' appears in a string
sample_text = "A random sentence with a few a's scattered around."
count_a = 0
for char in sample_text.lower():
    if char == 'a':
        count_a += 1
print(f"\nNumber of times 'a' appears: {count_a}")

# 7. Iterate over string and skip spaces using continue
print("\nCharacters in string without spaces:")
for ch in sample_text:
    if ch == ' ':
        continue
    print(ch)
```

```python
import sys

# 1. List of tuples – Print names with score > 75
scores = [("Alice", 88), ("Bob", 72), ("Charlie", 95), ("Diana", 60)]
print("People with scores greater than 75:")
for name, score in scores:
    if score > 75:
        print(name)

# 2. While loop asking for input until user types "exit"
print("\nType 'exit' to stop.")
while True:
    user_input = input("Enter something: ")
    if user_input.lower() == "exit":
        break
    print("You entered:", user_input)

# 3. Dictionary from 1 to 10 mapping to squares; print as set
square_dict = {}
for i in range(1, 11):
    square_dict[i] = i ** 2

square_set = set(square_dict.values())
print("\nSquares from 1 to 10 (as set):", square_set)

# 4. Print element pairs from two lists
list1 = ["a", "b", "c", "d"]
list2 = [10, 20, 30, 40]

print("\nPairs from both lists:")
for i in range(len(list1)):
    print(f"{list1[i]} - {list2[i]}")

# 5. Print pattern using nested loops
print("\nPattern:")
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

# 6. Square if even, cube if odd
original_numbers = [2, 5, 7, 8, 10, 11]
processed_numbers = []

for num in original_numbers:
    if num % 2 == 0:
        processed_numbers.append(num ** 2)
    else:
        processed_numbers.append(num ** 3)

print("\nOriginal numbers:", original_numbers)
print("Processed numbers (even squared, odd cubed):", processed_numbers)
```
